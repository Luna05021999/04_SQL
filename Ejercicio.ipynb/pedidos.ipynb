## Ejercicio 6

Obtener por cada mes (con el formato de nombre y no de número ej. Noviembre),la cantidad de pedidos realizados.

```python
connection = mysql.connector.connect(host=' ', database='  ', user='  ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query6 = """ SELECT MONTHNAME(Fecha) AS Mes, COUNT(numero) AS Cantidad_pedidos FROM PEDIDO GROUP BY MONTH(Fecha), monthname(Fecha) ORDER BY MONTH(Fecha); """
cursor.execute(query6)
cursor.fetchall()
tablas = "SHOW TABLES;"
df6 = pd.read_sql(query6, con=connection)
df6
```

## Ejercicio 7

Obtener un listado con el nombre y dni de los empleados en el mismo campo y anteponiendo el dni al nombre de la siguiente forma (ej; 45776633P, Juan Rodríguez López) que tenga turno de “tarde” o "noche". Ordenar el listado por dni.

```python
connection = mysql.connector.connect(host=' ', database='  ', user='  ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query7 = """ SELECT CONCAT(DNI, nombre) AS Empleado FROM EMPLEADO WHERE turno IN ('tarde', 'noche') ORDER BY DNI; """
cursor.execute(query7)
cursor.fetchall()
tablas = "SHOW TABLES;"
df7 = pd.read_sql(query7, con=connection)
df7
```

## Ejercicio 8

Obtener el nombre, código y precio de aquellos productos que superen o igualen la media de todos los precios. Ordenar de mayor a menor precio.

```python
connection = mysql.connector.connect(host=' ', database='  ', user='  ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query8 = """ SELECT codigo, nombre, precio AS Superan_media FROM PRODUCTO WHERE precio >= (SELECT AVG(Precio) FROM PRODUCTO) ORDER BY Precio DESC; """
cursor.execute(query8)
cursor.fetchall()
tablas = "SHOW TABLES;"
df8 = pd.read_sql(query8, con=connection)
df8
```
